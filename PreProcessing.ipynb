```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/.config/Bengaluru_House_Data.csv')
df.head()
```

```python
df.head()
```

```python
df.isnull()
null_counts=df.isnull().sum()
NaN_counts=df.isna().sum()
print(null_counts)
```

# New Section

```python
df1=df.copy()
df = df.dropna(subset=['location'])
df['location'].isnull().sum()
```

```python
df['size'] = df['size'].astype(str)
df['size'] = df['size'].str.extract('(\d+)').astype(float)
df['size'].head()
```

```python
df['size']=df['size'].fillna(df['size'].median())
df['size'].isnull().sum()
```

```python
df['society']=df['society'].fillna(method='ffill')
df['society'].isnull().sum()
```

```python
df['balcony']=df['balcony'].fillna(df['balcony'].mean())
df['balcony'].isnull().sum()
```

```python
df['bath']=df['bath'].fillna(df['bath'].mode())
df['bath']=df['bath'].replace(to_replace=np.nan,value=df['bath'].mode()[0])
df.isnull().sum()
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['area_type'] = label_encoder.fit_transform(df['area_type'])
df['location'] = label_encoder.fit_transform(df['location'])
df['availability'] = label_encoder.fit_transform(df['availability'])
df['society'] = label_encoder.fit_transform(df['society'])
df.head()
```

```python
NaN_counts=df.isna().sum()
print(NaN_counts)
```

```python
df=df[pd.to_numeric(df['total_sqft'],errors='coerce').notna()]
```
